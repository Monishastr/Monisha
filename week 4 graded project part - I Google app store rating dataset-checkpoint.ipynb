# 8.Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation)

```python
# Step 1: Convert the 'Reviews' column to a numeric data type
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Step 2: Calculate the IQR (Interquartile Range) for the 'Reviews' column
Q1 = df['Reviews'].quantile(0.25)
Q3 = df['Reviews'].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Define the IQR threshold 
IQR_threshold = 1.5

# Step 4: Identify outliers based on the IQR
outliers_low = df['Reviews'] < (Q1 - IQR_threshold * IQR)
outliers_high = df['Reviews'] > (Q3 + IQR_threshold * IQR)

# Step 5: Apply log transformation to handle outliers
df.loc[outliers_low, 'Reviews'] = np.log1p(df.loc[outliers_low, 'Reviews'])
df.loc[outliers_high, 'Reviews'] = np.log1p(df.loc[outliers_high, 'Reviews'])

# Step 6: Check if there are any remaining outliers
outliers = df[outliers_low | outliers_high]

if outliers.empty:
    print('No outliers found')
else:
    print("Outliers found and treated:", len(outliers))
    
```

# 9.The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries wheresize='Varies with device')
